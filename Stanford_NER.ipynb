```python
import stanza

# Download the English model if not already downloaded
stanza.download('en')

# Initialize the NLP pipeline
nlp = stanza.Pipeline('en')

# Sample documents
documents = [
    "Barack Obama was the 44th President of the United States.",
    "Microsoft Corporation is a technology company based in Redmond, Washington.",
    "Paris is the capital of France."
]

# Process each document and extract named entities
for i, doc in enumerate(documents, 1):
    print(f"Document {i}:")
    processed_doc = nlp(doc)
    for sentence in processed_doc.sentences:
        for entity in sentence.ents:
            print(f"  - Entity: '{entity.text}', Type: {entity.type}")
    print()
```
